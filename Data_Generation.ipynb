```python
equipment_names = [
    'Land-based Rotary Rig', 'Jack-up Rig', 'Semi-submersible Rig', 'Drillship',
    'PDC Drill Bit', 'Tricone Drill Bit', 'Mud Motor', 'Mud Pump',
    'Shale Shaker', 'Desander', 'Degasser', 'Blowout Preventer',
    'Top Drive', 'Casing Head', 'Cementing Unit', 'Coiled Tubing Unit',
    'Workover Rig', 'Wireline Unit', 'Subsea Tree', 'Christmas Tree',
    'Wellhead', 'Flowline', 'Separator', 'Heater Treater', 'Storage Tank',
    'Gas Compressor', 'Gas Dehydrator', 'Well Control Panel', 'Accumulator Unit',
    'Annular Preventer', 'Choke Manifold', 'Kill Line', 'Standpipe', 'Drill Collar',
    'Kelly Drive', 'Rotary Table', 'Draw Works', 'Swivel', 'Crown Block',
    'Travelling Block', 'Derrick', 'Mud Gas Separator', 'Cuttings Dryer', 'Desilter',
    'Trip Tank', 'Cement Silo', 'Surge Tank', 'Mixing Hopper', 'Well Logging Tool',
    'Formation Tester'
]

def generate_equipment():
    return [{'ID': i, 'Name': equipment_names[i-1]} for i in range(1, len(equipment_names)+1)]

equipment_specs = pd.DataFrame([
    {'Equipment': 'Land-based Rotary Rig', 'Category': 'Rig',             'Brand': 'RigCo',        'Model': 'LR-100',      'Max Depth Rating (m)': 5000, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 300, 'Hook Load Capacity (t)': 200, 'Daily Rate (USD)': 50000, 'Lead Time (days)': 14, 'In Stock': True},
    {'Equipment': 'Jack-up Rig',           'Category': 'Rig',             'Brand': 'OffshoreInc',  'Model': 'JU-45',       'Max Depth Rating (m)': 300,  'Max Pressure Rating (bar)': 350, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 250, 'Hook Load Capacity (t)': 150, 'Daily Rate (USD)': 60000, 'Lead Time (days)': 21, 'In Stock': False},
    {'Equipment': 'Semi-submersible Rig',  'Category': 'Rig',             'Brand': 'SubSeaWorks',  'Model': 'SSR-300',      'Max Depth Rating (m)': 1500, 'Max Pressure Rating (bar)': 450, 'Max Temperature Rating (°C)': 170, 'Torque Capacity (kN·m)': 280, 'Hook Load Capacity (t)': 180, 'Daily Rate (USD)': 75000, 'Lead Time (days)': 10, 'In Stock': True},
    {'Equipment': 'Drillship',             'Category': 'Rig',             'Brand': 'OceanDrill',   'Model': 'DS-900',       'Max Depth Rating (m)': 3500, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 190, 'Torque Capacity (kN·m)': 320, 'Hook Load Capacity (t)': 220, 'Daily Rate (USD)': 95000, 'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'PDC Drill Bit',         'Category': 'Drill Bit',       'Brand': 'BitCo',        'Model': 'PDCX-200',     'Max Depth Rating (m)': 4000, 'Max Pressure Rating (bar)': 300, 'Max Temperature Rating (°C)': 200, 'Torque Capacity (kN·m)': 150, 'Hook Load Capacity (t)':  50, 'Daily Rate (USD)': 1200,  'Lead Time (days)': 5,  'In Stock': True},
    {'Equipment': 'Tricone Drill Bit',     'Category': 'Drill Bit',       'Brand': 'RockDrill',    'Model': 'TRC-150',      'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 250, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 140, 'Hook Load Capacity (t)':  45, 'Daily Rate (USD)': 1000,  'Lead Time (days)': 3,  'In Stock': True},
    {'Equipment': 'Mud Motor',             'Category': 'Motor',           'Brand': 'FlowTech',     'Model': 'MM-50',        'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 200, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)':  90, 'Hook Load Capacity (t)':  30, 'Daily Rate (USD)': 800,   'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Mud Pump',              'Category': 'Pump',            'Brand': 'PumpMaster',   'Model': 'MP-300',       'Max Depth Rating (m)': 2500, 'Max Pressure Rating (bar)': 350, 'Max Temperature Rating (°C)': 150, 'Torque Capacity (kN·m)': 110, 'Hook Load Capacity (t)':  40, 'Daily Rate (USD)': 2000,  'Lead Time (days)': 10, 'In Stock': True},
    {'Equipment': 'Shale Shaker',          'Category': 'Separator',       'Brand': 'ShakePro',     'Model': 'SS-25',        'Max Depth Rating (m)': 1000, 'Max Pressure Rating (bar)': 100, 'Max Temperature Rating (°C)': 120, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 500,   'Lead Time (days)': 2,  'In Stock': True},
    {'Equipment': 'Desander',              'Category': 'Separator',       'Brand': 'CleanFlow',    'Model': 'DND-20',       'Max Depth Rating (m)': 1500, 'Max Pressure Rating (bar)': 200, 'Max Temperature Rating (°C)': 130, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 400,   'Lead Time (days)': 2,  'In Stock': True},
    {'Equipment': 'Degasser',              'Category': 'Separator',       'Brand': 'DegasTech',    'Model': 'DG-10',        'Max Depth Rating (m)': 1500, 'Max Pressure Rating (bar)': 150, 'Max Temperature Rating (°C)': 140, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 300,   'Lead Time (days)': 2,  'In Stock': True},
    {'Equipment': 'Blowout Preventer',     'Category': 'Safety',          'Brand': 'SafeWell',     'Model': 'BOP-500',      'Max Depth Rating (m)': 3500, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 250, 'Hook Load Capacity (t)': 100, 'Daily Rate (USD)': 3000,  'Lead Time (days)': 14, 'In Stock': True},
    {'Equipment': 'Top Drive',             'Category': 'Drive System',    'Brand': 'DriveMax',     'Model': 'TD-100',       'Max Depth Rating (m)': 4000, 'Max Pressure Rating (bar)': 400, 'Max Temperature Rating (°C)': 170, 'Torque Capacity (kN·m)': 180, 'Hook Load Capacity (t)':  80, 'Daily Rate (USD)': 2500,  'Lead Time (days)': 10, 'In Stock': True},
    {'Equipment': 'Casing Head',           'Category': 'Wellhead',        'Brand': 'HeadSafe',     'Model': 'CH-75',        'Max Depth Rating (m)': 4500, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 160, 'Hook Load Capacity (t)':  70, 'Daily Rate (USD)': 1200,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Cementing Unit',        'Category': 'Cementing',       'Brand': 'CemenTech',    'Model': 'CU-80',        'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 300, 'Max Temperature Rating (°C)': 150, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1500,  'Lead Time (days)': 5,  'In Stock': True},
    {'Equipment': 'Coiled Tubing Unit',    'Category': 'Intervention',     'Brand': 'TubeWorks',    'Model': 'CTU-200',      'Max Depth Rating (m)': 3500, 'Max Pressure Rating (bar)': 250, 'Max Temperature Rating (°C)': 140, 'Torque Capacity (kN·m)': 100, 'Hook Load Capacity (t)':  40, 'Daily Rate (USD)': 4500,  'Lead Time (days)': 10, 'In Stock': False},
    {'Equipment': 'Workover Rig',          'Category': 'Rig',             'Brand': 'ReRig',        'Model': 'WR-50',        'Max Depth Rating (m)': 2500, 'Max Pressure Rating (bar)': 350, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 140, 'Hook Load Capacity (t)':  60, 'Daily Rate (USD)': 40000, 'Lead Time (days)': 21, 'In Stock': True},
    {'Equipment': 'Wireline Unit',         'Category': 'Intervention',     'Brand': 'LineTech',     'Model': 'WL-30',        'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 200, 'Max Temperature Rating (°C)': 150, 'Torque Capacity (kN·m)':  60, 'Hook Load Capacity (t)':  20, 'Daily Rate (USD)': 3500,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Subsea Tree',           'Category': 'Wellhead',        'Brand': 'SubSeaHead',   'Model': 'ST-100',       'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 200, 'Torque Capacity (kN·m)': 200, 'Hook Load Capacity (t)': 120, 'Daily Rate (USD)': 6000,  'Lead Time (days)': 14, 'In Stock': False},
    {'Equipment': 'Christmas Tree',        'Category': 'Wellhead',        'Brand': 'XmasOil',      'Model': 'CT-200',       'Max Depth Rating (m)': 2800, 'Max Pressure Rating (bar)': 550, 'Max Temperature Rating (°C)': 190, 'Torque Capacity (kN·m)': 180, 'Hook Load Capacity (t)': 110, 'Daily Rate (USD)': 6200,  'Lead Time (days)': 14, 'In Stock': True},
    {'Equipment': 'Wellhead',              'Category': 'Wellhead',        'Brand': 'HeadTech',     'Model': 'WH-50',        'Max Depth Rating (m)': 4000, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 150, 'Hook Load Capacity (t)':  75, 'Daily Rate (USD)': 3000,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Flowline',              'Category': 'Pipework',        'Brand': 'FlowPro',      'Model': 'FL-20',        'Max Depth Rating (m)': 1000, 'Max Pressure Rating (bar)': 200, 'Max Temperature Rating (°C)': 120, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 800,   'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Separator',             'Category': 'Separator',       'Brand': 'SeparateIt',   'Model': 'SP-100',       'Max Depth Rating (m)': 1500, 'Max Pressure Rating (bar)': 250, 'Max Temperature Rating (°C)': 140, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1200,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Heater Treater',        'Category': 'Separator',       'Brand': 'HeatTreat',    'Model': 'HT-300',       'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 300, 'Max Temperature Rating (°C)': 200, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1500,  'Lead Time (days)': 5,  'In Stock': True},
    {'Equipment': 'Storage Tank',          'Category': 'Storage',         'Brand': 'TankStore',    'Model': 'STK-500',      'Max Depth Rating (m)': 0,    'Max Pressure Rating (bar)': 50,  'Max Temperature Rating (°C)': 80,  'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1000,  'Lead Time (days)': 14, 'In Stock': True},
    {'Equipment': 'Gas Compressor',        'Category': 'Compressor',      'Brand': 'CompAir',      'Model': 'GC-150',       'Max Depth Rating (m)': 2500, 'Max Pressure Rating (bar)': 400, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 100, 'Hook Load Capacity (t)':  30, 'Daily Rate (USD)': 2800,  'Lead Time (days)': 7,  'In Stock': False},
    {'Equipment': 'Gas Dehydrator',        'Category': 'Gas Treating',    'Brand': 'DryGas',       'Model': 'GD-200',       'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 300, 'Max Temperature Rating (°C)': 140, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 2200,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Well Control Panel',    'Category': 'Control',         'Brand': 'ControlSys',   'Model': 'WCP-50',       'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 450, 'Max Temperature Rating (°C)': 170, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1800,  'Lead Time (days)': 10, 'In Stock': True},
    {'Equipment': 'Accumulator Unit',      'Category': 'Hydraulic',       'Brand': 'AccuTech',     'Model': 'ACU-30',       'Max Depth Rating (m)': 2500, 'Max Pressure Rating (bar)': 350, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 2000,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Annular Preventer',     'Category': 'Safety',          'Brand': 'SafeWell',     'Model': 'AP-200',       'Max Depth Rating (m)': 3500, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 3200,  'Lead Time (days)': 14, 'In Stock': False},
    {'Equipment': 'Choke Manifold',        'Category': 'Safety',          'Brand': 'ChokePro',     'Model': 'CM-100',       'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 550, 'Max Temperature Rating (°C)': 170, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 2800,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Kill Line',            'Category': 'Safety',          'Brand': 'KillSafe',     'Model': 'KL-50',        'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 1500,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Standpipe',            'Category': 'Pipework',        'Brand': 'PipeFlow',     'Model': 'SP-40',        'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 400, 'Max Temperature Rating (°C)': 150, 'Torque Capacity (kN·m)':   0, 'Hook Load Capacity (t)':   0, 'Daily Rate (USD)': 900,   'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Drill Collar',         'Category': 'Drill String',    'Brand': 'DrillCo',      'Model': 'DC-30',        'Max Depth Rating (m)': 4500, 'Max Pressure Rating (bar)': 500, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 250, 'Hook Load Capacity (t)':  80, 'Daily Rate (USD)': 1800,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Kelly Drive',          'Category': 'Drive System',    'Brand': 'DriveCo',      'Model': 'KD-20',        'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 350, 'Max Temperature Rating (°C)': 160, 'Torque Capacity (kN·m)': 150, 'Hook Load Capacity (t)':  50, 'Daily Rate (USD)': 1400,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Rotary Table',         'Category': 'Drive System',    'Brand': 'TableTech',    'Model': 'RT-50',        'Max Depth Rating (m)': 2000, 'Max Pressure Rating (bar)': 300, 'Max Temperature Rating (°C)': 150, 'Torque Capacity (kN·m)': 140, 'Hook Load Capacity (t)':  45, 'Daily Rate (USD)': 1300,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Draw Works',           'Category': 'Hoisting',        'Brand': 'HoistPro',     'Model': 'DW-100',       'Max Depth Rating (m)': 5000, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 300, 'Hook Load Capacity (t)': 200, 'Daily Rate (USD)': 2200,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Swivel',               'Category': 'Hoisting',        'Brand': 'SwivelTech',   'Model': 'SV-20',        'Max Depth Rating (m)': 3000, 'Max Pressure Rating (bar)': 400, 'Max Temperature Rating (°C)': 170, 'Torque Capacity (kN·m)': 180, 'Hook Load Capacity (t)':  60, 'Daily Rate (USD)': 1600,  'Lead Time (days)': 7,  'In Stock': True},
    {'Equipment': 'Crown Block',          'Category': 'Hoisting',        'Brand': 'BlockWorks',   'Model': 'CB-60',        'Max Depth Rating (m)': 5000, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)': 250, 'Hook Load Capacity (t)': 150, 'Daily Rate (USD)': 3000,  'Lead Time (days)': 10, 'In Stock': True},
    {'Equipment': 'Travelling Block',     'Category': 'Hoisting',        'Brand': 'BlockWorks',   'Model': 'TB-60',        'Max Depth Rating (m)': 5000, 'Max Pressure Rating (bar)': 600, 'Max Temperature Rating (°C)': 180, 'Torque Capacity (kN·m)':250, 'Hook Load Capacity (t)':150, 'Daily Rate (USD)':3000, 'Lead Time (days)':10,'In Stock':True},
    {'Equipment': 'Derrick',              'Category': 'Structure',       'Brand': 'DerrickCo',    'Model': 'DR-80',        'Max Depth Rating (m)':5000, 'Max Pressure Rating (bar)':600, 'Max Temperature Rating (°C)':180, 'Torque Capacity (kN·m)':0, 'Hook Load Capacity (t)':0, 'Daily Rate (USD)':3500, 'Lead Time (days)':15,'In Stock':True},
    {'Equipment': 'Mud Gas Separator',    'Category': 'Safety',          'Brand': 'GasSafe',      'Model': 'MGS-30',       'Max Depth Rating (m)':2000,'Max Pressure Rating (bar)':300,'Max Temperature Rating (°C)':150,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':1800,'Lead Time (days)':7,'In Stock':True},
    {'Equipment': 'Cuttings Dryer',       'Category': 'Separator',       'Brand': 'DryPro',       'Model': 'CD-40',        'Max Depth Rating (m)':1500,'Max Pressure Rating (bar)':200,'Max Temperature Rating (°C)':140,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':1000,'Lead Time (days)':5,'In Stock':True},
    {'Equipment': 'Desilter',             'Category': 'Separator',       'Brand': 'SilPro',       'Model': 'DSR-30',       'Max Depth Rating (m)':1500,'Max Pressure Rating (bar)':200,'Max Temperature Rating (°C)':140,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':900,'Lead Time (days)':5,'In Stock':True},
    {'Equipment': 'Trip Tank',            'Category': 'Fluid Control',   'Brand': 'TripFlow',     'Model': 'TT-20',        'Max Depth Rating (m)':2000,'Max Pressure Rating (bar)':250,'Max Temperature Rating (°C)':150,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':800,'Lead Time (days)':3,'In Stock':True},
    {'Equipment': 'Cement Silo',          'Category': 'Cementing',       'Brand': 'CemenStore',   'Model': 'CS-100',       'Max Depth Rating (m)':0,   'Max Pressure Rating (bar)':50, 'Max Temperature Rating (°C)':80, 'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':700,'Lead Time (days)':5,'In Stock':True},
    {'Equipment': 'Surge Tank',           'Category': 'Fluid Control',   'Brand': 'SurgeSafe',    'Model': 'ST-30',        'Max Depth Rating (m)':0,   'Max Pressure Rating (bar)':100,'Max Temperature Rating (°C)':120,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':600,'Lead Time (days)':5,'In Stock':True},
    {'Equipment': 'Mixing Hopper',        'Category': 'Cementing',       'Brand': 'MixPro',       'Model': 'MH-25',        'Max Depth Rating (m)':0,   'Max Pressure Rating (bar)':50, 'Max Temperature Rating (°C)':100,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':400,'Lead Time (days)':3,'In Stock':True},
    {'Equipment': 'Well Logging Tool',    'Category': 'Logging',         'Brand': 'LogTech',      'Model': 'WL-10',        'Max Depth Rating (m)':5000,'Max Pressure Rating (bar)':400,'Max Temperature Rating (°C)':200,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':5000,'Lead Time (days)':14,'In Stock':True},
    {'Equipment': 'Formation Tester',     'Category': 'Logging',         'Brand': 'TestPro',      'Model': 'FT-20',        'Max Depth Rating (m)':5000,'Max Pressure Rating (bar)':500,'Max Temperature Rating (°C)':200,'Torque Capacity (kN·m)':0,'Hook Load Capacity (t)':0,'Daily Rate (USD)':5500,'Lead Time (days)':14,'In Stock':False}
])

NON_FAILURE_SYMPTOMS = [
    'Abnormal vibration','Pressure loss','Lack of pressure',
    'Fluid leakage','Oil flow anomaly'
]
ALL_SYMPTOMS = NON_FAILURE_SYMPTOMS + ['Stopped working']

# Date generator
def generate_dates(start, end):
    return pd.date_range(start, end, freq='D')

# Water zone helper
def _assign_zone(loc, wd):
    if loc == 'Onshore': return 'Onshore'
    if wd < 100: return 'Negligible'
    if wd < 300: return 'Shallow'
    if wd < 600: return 'Mid-shallow'
    if wd < 1000: return 'Mid'
    if wd < 1500: return 'Mid-deep'
    if wd < 2000: return 'Deep'
    return 'Very deep'

def generate_data(start: str, end: str, min_per_day: int = 5, max_per_day: int = 15) -> pd.DataFrame:
    rows = []
    equips = [e['Name'] for e in generate_equipment()]

    for day in generate_dates(start, end):
        for _ in range(random.randint(min_per_day, max_per_day)):
            ts = day + pd.Timedelta(seconds=random.random()*86400)
            eq = random.choice(equips)
            spec = equipment_specs.loc[equipment_specs['Equipment']==eq].iloc[0]

            loc = random.choice(['Onshore','Offshore'])
            wd = random.uniform(50,2000) if loc=='Offshore' else np.nan
            zone = _assign_zone(loc, wd)
            formation = random.choice(['Sandstone','Shale','Limestone','Granite','Dolomite','Basalt','Soft clay'])
            depth_m = random.uniform(500,6000)
            formation_pres = random.uniform(50,500)
            mud_wt, mud_visc = random.uniform(8.5,18), random.uniform(10,40)
            pump_pres, pump_flow = random.uniform(20,100), random.uniform(200,600)
            rop = random.uniform(5,60)
            hook_load = random.uniform(100,500)
            torque = random.uniform(50,300)
            maint = random.choice(['Preventive','Corrective'])
            temp = random.uniform(70,130)
            pres = random.uniform(8,16)
            vib = random.uniform(1,3)
            hrs = random.randint(16,24)
            parts = random.sample(['Bearing','Valve','Filter','Heat Exchanger','Compressor'], k=random.randint(1,3))
            replaced_parts = ', '.join(parts)
            fail_cause = random.choice(['Natural wear','Electrical failure','Leakage','Mechanical issue'])
            comp_part = random.choice(['Bearing','Oil Filter','Pressure Valve','Heat Exchanger','Flow Control Valve'])

            base_flags = [
                depth_m>3000, formation_pres>300, formation in ['Shale','Basalt'],
                mud_wt<9 or mud_wt>17, mud_visc>35, pump_pres>80,
                pump_flow<250 or pump_flow>550, rop>50, hook_load>450,
                torque>250, maint=='Corrective', temp>110, pres<9,
                vib>2.5, hrs>22, len(parts)>2,
                fail_cause in ['Electrical failure','Leakage'],
                comp_part in ['Bearing','Pressure Valve']
            ]
            spec_flags = [
                depth_m>spec['Max Depth Rating (m)'],
                pres>spec['Max Pressure Rating (bar)'],
                temp>spec['Max Temperature Rating (°C)'],
                torque>spec['Torque Capacity (kN·m)'],
                hook_load>spec['Hook Load Capacity (t)']
            ]
            score = sum(base_flags)+sum(spec_flags)
            p_fail = 1/(1+np.exp(-(score-5)/3))
            failure = int(random.random()<p_fail)
            symptom = (random.choices(ALL_SYMPTOMS, k=1)[0] if failure else random.choices(NON_FAILURE_SYMPTOMS, k=1)[0])

            rows.append({
                'Timestamp':ts,'Equipment':eq,'Category':spec['Category'],'Brand':spec['Brand'],
                'Model':spec['Model'],'Location Type':loc,'Water Zone':zone,'Formation Type':formation,
                'Drilling Depth (m)':depth_m,'Formation Pressure (bar)':formation_pres,
                'Mud Weight (ppg)':mud_wt,'Mud Viscosity (cP)':mud_visc,
                'Pump Pressure (bar)':pump_pres,'Pump Flow (L/min)':pump_flow,
                'ROP (m/hr)':rop,'Hook Load (t)':hook_load,'Torque (kN·m)':torque,
                'Maintenance Type':maint,'Temperature (°C)':temp,'Pressure (bar)':pres,
                'Vibration (mm/s)':vib,'Operating Hours':hrs,'Replaced Parts':replaced_parts,
                'Failure Cause':fail_cause,'Part':comp_part,'Daily Rate (USD)':spec['Daily Rate (USD)'],
                'Lead Time (days)':spec['Lead Time (days)'],'In Stock':spec['In Stock'],
                'Failure Class':failure,'Observed Symptom':symptom
            })

    df = pd.DataFrame(rows)
    df0, df1 = df[df['Failure Class']==0], df[df['Failure Class']==1]
    n = min(len(df0),len(df1))
    df_bal = pd.concat([df0.sample(n,random_state=42), df1.sample(n,random_state=42)])
    return df_bal.sample(frac=1,random_state=42).reset_index(drop=True)
```
